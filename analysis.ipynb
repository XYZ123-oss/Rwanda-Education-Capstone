# Rwanda Education Capstone Analysis

This notebook performs an exploratory analysis of Rwandan education enrollment data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('../data/rwanda_education_dataset.csv')
df.head()
```

## Total Enrollment by Year

```python
sns.barplot(data=df, x='Year', y='Enrollment', estimator=sum, ci=None)
plt.title('Total Enrollment by Year')
plt.show()
```

## Enrollment by Gender

```python
sns.barplot(data=df, x='Gender', y='Enrollment', estimator=sum, ci=None)
plt.title('Total Enrollment by Gender')
plt.show()
```
